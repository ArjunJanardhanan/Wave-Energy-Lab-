```python
import matplotlib.pyplot as plt

plt.plot(df['frequency'], df['average_amplitude_f2'], marker='o', linestyle='--')
plt.xlabel('Wave frequency (Hz)')  # Set the label for the X-axis
plt.ylabel('Amplitude incident wave (m)') # Set the label for the Y-axis
plt.title('Plot of amplitude of incident wave vs Wave frequency')  # Set the title of the plot
plt.grid(True)  # Show gridlines
plt.show()  # Display the plot
```
